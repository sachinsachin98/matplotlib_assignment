#Q1
Matplotlib is a popular Python library used for creating visualizations and plots. It provides a wide range of tools for creating various types of plots and charts to visualize data in a meaningful way. Matplotlib is extensively used in scientific research, data analysis, and other fields where data visualization is crucial.

Five types of plots that can be created using the Pyplot module of Matplotlib:

1)Line Plot: A line plot is used to visualize the relationship between two variables by connecting data points with lines. It's commonly used to show trends or patterns over time.

2)Scatter Plot: A scatter plot displays individual data points as dots, allowing you to examine the relationship between two variables and identify any potential correlations.

3)Bar Plot: A bar plot, also known as a bar chart, represents categorical data using rectangular bars. It's useful for comparing the quantities or frequencies of different categories.

4)Histogram: A histogram is used to display the distribution of a continuous dataset. It groups data into bins and shows how many data points fall into each bin, providing insights into the data's distribution.

5)Pie Chart: A pie chart is a circular chart divided into sectors, where each sector represents a proportion of a whole. It's suitable for displaying parts of a whole and comparing relative sizes.

```python
#Q2
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,6))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("scatter plot")
plt.grid()
```

Scatter plots are tool to draw relationships between objects using dots to represent respective values

#Q3
The subplot() function in Matplotlib is used to create multiple plots within a single figure, arranged in a grid-like fashion. This is particularly useful when you want to compare or display multiple plots side by side. The subplot() function takes three integer arguments: the number of rows, the number of columns, and the index of the current subplot.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
 
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
 
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
plt.figure(figsize=(4,4))

plt.subplot(2,2,1)
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2,2,2)
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2,2,3)
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2,2,4)
plt.plot(x4, y4)
plt.title("Line 1")

plt.show()
```

#Q4
A bar plot, also known as a bar chart, is a type of data visualization that uses rectangular bars to represent data. Each bar corresponds to a category, and the length of the bar is proportional to the value it represents. Bar plots are commonly used to compare the values of different categories and show their relationships.

Bar plots are used to:

Compare Data: Bar plots are effective in visually comparing the values of different categories, making it easy to identify which category has the highest or lowest value.
Show Distribution: They can be used to display the distribution of a dataset across different categories.
Highlight Trends: Bar plots can highlight trends and patterns within the data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(4, 4))
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits')
plt.show()

plt.figure(figsize=(4, 4))
plt.barh(company, profit)
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits')
plt.show()
```

#Q5
A box plot, also known as a box-and-whisker plot, is a type of data visualization that provides a concise summary of the distribution of a dataset. It displays the median, quartiles, and potential outliers of the data in a compact format. A box plot is particularly useful for identifying the central tendency, spread, and potential skewness or outliers in a dataset.

A box plot is used to:

Visualize Data Spread: It shows the range of data and the interquartile range, providing insights into the spread or dispersion of the dataset.
Identify Outliers: Outliers, which are data points significantly different from the majority of the dataset, are shown as individual points outside the "whiskers."
Compare Distributions: Box plots are helpful for comparing the distribution of data across different categories or groups.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data = [box1, box2]
labels = ['box1', 'box2']
plt.figure(figsize=(4, 4))
plt.boxplot(data, labels=labels)
plt.xlabel('boxes')
plt.ylabel('values')
plt.title('box-plot')
plt.show()
```
